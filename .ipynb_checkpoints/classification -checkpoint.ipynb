# FLAML

Now we will import flaml  and from Flaml we will import Automl 

```python
import flaml as fl
```

```python
from flaml import AutoML
```

```python
automl=AutoML()
```

Now let's import the iris data from the sklearn 

```python
from sklearn.datasets import load_iris   
```

```python
X_train, y_train = load_iris(return_X_y=True)  #now i will train the the dta into X_train and y_train
```

```python
automl.fit(X_train, y_train, task="classification") # Now we will fit the data and use the Classification 
```

```python
print(automl.predict_proba(X_train)) #Now the Training is done so lets Perdict the the values
```

```python
print(automl.model.estimator) # lets now the best model Estimator 
```
